```python
from sklearn.decomposition import PCA

# visualize data cluster using PCA
# convert to numpy 
timestamp_cols = data.columns[0]
feature_cols = data.columns[1:-1]
condition_cols = data.columns[-1]
timestamps = data[timestamp_cols].values
features = data[feature_cols].values
labels = data[condition_cols].values
print(features)
print(labels)
```

```python
def split_data(data, labels):
    pos_data = []
    neg_data = []
    for i, d in enumerate(data):
        if labels[i] == 1:
            pos_data.append(d)
        elif labels[i] == 0:
            neg_data.append(d)
    return np.array(pos_data), np.array(neg_data)
```

```python
# averaging data of two chiller pump
CHP1Temp = np.mean([data['CHP1Temp1(Deg C)'].values, data['CHP1Temp2(Deg C)'].values],axis=0)
CHP2Temp = np.mean([data['CHP2Temp1(Deg C)'].values, data['CHP2Temp2(Deg C)'].values],axis=0)
CHP1Vib = np.mean([data['CHP1Vib1(mm/s)'].values, data['CHP1Vib2(mm/s)'].values],axis=0)
CHP2Vib = np.mean([data['CHP2Vib1(mm/s)'].values, data['CHP2Vib2(mm/s)'].values],axis=0)
temp = [CHP1Temp, CHP2Temp]
vib = [CHP1Vib, CHP2Vib]
# visualize temp and vib data of two chiller pump
fig, ax = plt.subplots(1, 2, figsize=(13, 5))
for i in range(2):
    fault_temp, normal_temp = split_data(temp[i], labels)
    fault_vib, normal_vib = split_data(vib[i], labels)
    ax[i].scatter(normal_vib, normal_temp, label='normal')
    ax[i].scatter(fault_vib, fault_temp, label='fault',marker='x', s=30)
    ax[i].set_ylabel('Temp (Deg C)')
    ax[i].set_xlabel('Vib (mm/s)')
    ax[i].set_title(f'Chiller Pump {i+1}')
    ax[i].legend()


plt.show()
```

From this PCA visualization, there is clear disctinction between fault and normal pump. To make sure our assumption, We will gather more insight from sensor data that we have. There are 2 type of data, temperature and vibration. Each one of them is produced from 4 different sensors.  

```python
def moving_avg(data, w_size=50):
    avg_move = []
    for i in range(len(data)):
        if i - w_size < 0:
            mean = np.mean(data[0: i])
        else: 
            mean = np.mean(data[i - w_size: i])
        avg_move.append(mean)
    return np.array(avg_move)
```

```python
# visualize temperature data
t11 = data[feature_cols[0]]
t12 = data[feature_cols[1]]
t21 = data[feature_cols[2]]
t22 = data[feature_cols[3]]
tavg = np.mean([t11, t12, t21, t22], axis=0)

t11 = moving_avg(t11)
t12 = moving_avg(t12)
t21 = moving_avg(t21)
t22 = moving_avg(t22)
tavg = moving_avg(tavg)
```

```python
# visualize temperature data
v11 = data[feature_cols[4]]
v12 = data[feature_cols[5]]
v21 = data[feature_cols[6]]
v22 = data[feature_cols[7]]
vavg = np.mean([v11, v12, v21, v22], axis=0)

v11 = moving_avg(v11)
v12 = moving_avg(v12)
v21 = moving_avg(v21)
v22 = moving_avg(v22)
vavg = moving_avg(vavg)
```

```python
plt.xlim(0, len(t11))

for i in range(len(labels)):
    if labels[i] == 1:
        plt.axvspan(i, i+1, facecolor='r', alpha=0.1)
    elif labels[i] == 0:
        plt.axvspan(i, i+1, facecolor='g', alpha=0.1)

plt.plot(t11, label='t11')
plt.plot(t12, label='t12')
plt.plot(t21, label='t21')
plt.plot(t22, label='t22')
plt.plot(tavg, label='tavg')
plt.ylabel('Temp (Deg C)')
plt.xlabel('Time')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.xlim(0, len(v11))

for i in range(len(labels)):
    if labels[i] == 1:
        plt.axvspan(i, i+1, facecolor='r', alpha=0.1)
    elif labels[i] == 0:
        plt.axvspan(i, i+1, facecolor='g', alpha=0.1)
plt.plot(v11, label='v11')
plt.plot(v12, label='v12')
plt.plot(v21, label='v21')
plt.plot(v22, label='v22')
plt.plot(vavg, label='vavg')
plt.legend(loc='upper right')
plt.show()
```

From the data visualizations, we can imply that machine is in fault when the average temperature and vibration are lower than when it is normal. There is a sudden increase of temperature when there is vibrations, and also there is a sudden decrease of it when there is no vibration. Judging by the seasonality of the data, it is looks like there is two state of the machine. the first state is when there is no vibration, and second state when there is a vibration in the machine.

```python
from sklearn.model_selection import train_test_split
# Building train and test data for model
X = features
y = labels
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.1, stratify=y)
print(len(X_train))
print(len(X_test))
```

```python
# standardize data
from sklearn.preprocessing import StandardScaler

# we will differentiate scaler of vibration and temperature
scaler_temp = StandardScaler()
scaler_vib = StandardScaler()

scaler_temp.fit(X_train[:, 0:4])
scaler_vib.fit(X_train[:, 4:])

X_train[:, 0:4] = scaler_temp.transform(X_train[:, 0:4])
X_train[:, 4:] = scaler_vib.transform(X_train[:, 4:])
X_test[:, 0:4] = scaler_temp.transform(X_test[:, 0:4])
X_test[:, 4:] = scaler_vib.transform(X_test[:, 4:])
```

```python
# saving scaler for future prediction
import pickle

pickle.dump(scaler_temp, open('scaler_temp.pkl', 'wb'))
pickle.dump(scaler_vib, open('scaler_vib.pkl', 'wb'))
```

```python
# training model
from sklearn.svm import SVC

model = SVC(kernel='rbf')
model.fit(X_train, y_train)
acc = model.score(X_test, y_test)
print('acc:', acc)
```

```python
# saving model for another prediction
import pickle

pickle.dump(model, open('model_q1.pkl', 'wb'))
```

```python
# load model and scaler
model = pickle.load(open('model_q1.pkl', 'rb'))
scaler_temp = pickle.load(open('scaler_temp.pkl', 'rb'))
scaler_vib = pickle.load(open('scaler_vib.pkl', 'rb'))
```

```python
# try to predict from given input data
t11 = 17
t12 = 16
t21 = 12
t22 = 15

v11 = 0.5
v12 = 0.5
v21 = 0.6
v22 = 0.1

feature = np.array([[t11, t12, t21, t22, v11, v12, v21, v22]])
feature[:, 0:4] = scaler_temp.transform(feature[:, 0:4])
feature[:, 0:4] = scaler_vib.transform(feature[:, 4:])

y = model.predict(feature)
if y[0] == 0:
    print('prediction: normal')
elif y[0] == 1:
    print('prediction: fault')
```
