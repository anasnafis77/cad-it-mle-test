```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Q1.csv')
```

```python
from sklearn.decomposition import PCA

# visualize data cluster using PCA
# convert to numpy 
timestamp_cols = data.columns[0]
feature_cols = data.columns[1:-1]
condition_cols = data.columns[-1]
features = data[feature_cols].values
faulty = data[condition_cols].values
print(features)
print(faulty)
```

```python
# compress feature dimension into 2 principal component
pca = PCA(n_components=2,)
compressed_ft = pca.fit_transform(features)

# split compressed feature into fault and success
fault = []
success = []
for i, ft in enumerate(compressed_ft):
    if faulty[i] == 1:
        fault.append(list(ft))
    elif faulty[i] == 0:
        success.append(list(ft))
fault = np.array(fault)
success = np.array(success)

plt.scatter(fault[:, 0], fault[:, 1], label='fault')
plt.scatter(success[:, 0], success[:, 1], label='success')
plt.legend()
plt.show()
```

From this PCA visualization, there is clear disctinction between fault and success features through PCA. So we have a good start to make an assumption that we can build machine learning model to detect faulty of this machine using this data. To make sure our assumption, We will gather more insight from sensor data that we have. There are 2 type of data, temperature and vibration. Each one of them is produced from 4 different sensors.  

```python
def moving_avg(data, w_size=50):
    avg_move = []
    for i in range(len(data)):
        if i - w_size < 0:
            mean = np.mean(data[0: i])
        else: 
            mean = np.mean(data[i - w_size: i])
        avg_move.append(mean)
    return np.array(avg_move)
```

```python
# visualize temperature data
t1 = data[feature_cols[0]]
t2 = data[feature_cols[1]]
t3 = data[feature_cols[2]]
t4 = data[feature_cols[3]]
tavg = np.mean([t1, t2, t3, t4], axis=0)

t1 = moving_avg(t1)
t2 = moving_avg(t2)
t3 = moving_avg(t3)
t4 = moving_avg(t4)
tavg = moving_avg(tavg)
plt.xlim(0, len(t1))

for i in range(len(faulty)):
    if faulty[i] == 1:
        plt.axvspan(i, i+1, facecolor='r', alpha=0.1)
plt.plot(t1, label='t1')
plt.plot(t2, label='t2')
plt.plot(t3, label='t3')
plt.plot(t4, label='t4')
plt.plot(tavg, label='tavg')
plt.legend()
plt.show()
```

```python
# visualize temperature data
v1 = data[feature_cols[4]]
v2 = data[feature_cols[5]]
v3 = data[feature_cols[6]]
v4 = data[feature_cols[7]]
vavg = np.mean([v1, v2, v3, v4], axis=0)

v1 = moving_avg(v1)
v2 = moving_avg(v2)
v3 = moving_avg(v3)
v4 = moving_avg(v4)
vavg = moving_avg(vavg)
plt.xlim(0, len(v1))

for i in range(len(faulty)):
    if faulty[i] == 1:
        plt.axvspan(i, i+1, facecolor='r', alpha=0.1)
    elif faulty[i] == 0:
        plt.axvspan(i, i+1, facecolor='g', alpha=0.1)

plt.plot(v1, label='v1')
plt.plot(v2, label='v2')
plt.plot(v3, label='v3')
plt.plot(v4, label='v4')
plt.plot(vavg, label='vavg')
plt.legend()
plt.show()
```
